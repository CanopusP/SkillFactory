```python
Суть проекта — отследить влияние условий жизни учащихся в возрасте от 15 до 22 лет на их успеваемость по математике, чтобы на ранней стадии выявлять студентов, находящихся в группе риска.
1 school — аббревиатура школы, в которой учится ученик
2 sex — пол ученика ('F' - женский, 'M' - мужской)
3 age — возраст ученика (от 15 до 22)
4 address — тип адреса ученика ('U' - городской, 'R' - за городом)
5 famsize — размер семьи('LE3' <= 3, 'GT3' >3)
6 Pstatus — статус совместного жилья родителей ('T' - живут вместе 'A' - раздельно)
7 Medu — образование матери (0 - нет, 1 - 4 класса, 2 - 5-9 классы, 3 - среднее специальное или 11 классов, 4 - высшее)
8 Fedu — образование отца (0 - нет, 1 - 4 класса, 2 - 5-9 классы, 3 - среднее специальное или 11 классов, 4 - высшее)
9 Mjob — работа матери ('teacher' - учитель, 'health' - сфера здравоохранения, 'services' - гос служба, 'at_home' - не работает, 'other' - другое)
10 Fjob — работа отца ('teacher' - учитель, 'health' - сфера здравоохранения, 'services' - гос служба, 'at_home' - не работает, 'other' - другое)
11 reason — причина выбора школы ('home' - близость к дому, 'reputation' - репутация школы, 'course' - образовательная программа, 'other' - другое)
12 guardian — опекун ('mother' - мать, 'father' - отец, 'other' - другое)
13 traveltime — время в пути до школы (1 - <15 мин., 2 - 15-30 мин., 3 - 30-60 мин., 4 - >60 мин.)
14 studytime — время на учёбу помимо школы в неделю (1 - <2 часов, 2 - 2-5 часов, 3 - 5-10 часов, 4 - >10 часов)
15 failures — количество внеучебных неудач (n, если 1<=n<=3, иначе 0)
16 schoolsup — дополнительная образовательная поддержка (yes или no)
17 famsup — семейная образовательная поддержка (yes или no)
18 paid — дополнительные платные занятия по математике (yes или no)
19 activities — дополнительные внеучебные занятия (yes или no)
20 nursery — посещал детский сад (yes или no)
21 higher — хочет получить высшее образование (yes или no)
22 internet — наличие интернета дома (yes или no)
23 romantic — в романтических отношениях (yes или no)
24 famrel — семейные отношения (от 1 - очень плохо до 5 - очень хорошо)
25 freetime — свободное время после школы (от 1 - очень мало до 5 - очень мого)
26 goout — проведение времени с друзьями (от 1 - очень мало до 5 - очень много)
27 health — текущее состояние здоровья (от 1 - очень плохо до 5 - очень хорошо)
28 absences — количество пропущенных занятий
29 score — баллы по госэкзамену по математике
```

```python
data2.describe()
```

```python
# удаление выбросов по ср.баллу
perc25 = data2.score.quantile(0.25)
perc75 = data2.score.quantile(0.75)
IQR = data2.score.quantile(0.75) - data2.score.quantile(0.25)
score_data = data2[data2.score.between(perc25 - 1.5*IQR,
perc75 + 1.5*IQR)]
score_data.info()
```

```python
# перевести все пустые строки в формат строки "None"
score_data.address = score_data.address.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.famsize = score_data.famsize.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.Pstatus = score_data.Pstatus.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.Medu = score_data.Medu.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.Fedu = score_data.Fedu.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.Mjob = score_data.Mjob.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.Fjob = score_data.Fjob.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.reason = score_data.reason.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.guardian = score_data.guardian.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.traveltime = score_data.traveltime.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.studytime = score_data.studytime.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.failures = score_data.failures.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.schoolsup = score_data.schoolsup.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.famsup = score_data.famsup.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.paid = score_data.paid.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.activities = score_data.activities.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.nursery = score_data.nursery.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.higher = score_data.higher.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.internet = score_data.internet.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.romantic = score_data.romantic.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.famrel = score_data.famrel.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.freetime = score_data.freetime.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.goout = score_data.goout.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.health = score_data.health.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.absences = score_data.absences.apply(lambda x: 'None' if pd.isnull(x) else x)
score_data.info()
```

# absences

```python
absences_data = score_data[score_data.absences != 'None']
absences_data['absences'] = absences_data['absences'].astype(float)
perc25 = absences_data.absences.quantile(0.25)
perc75 = absences_data.absences.quantile(0.75)
IQR = absences_data.absences.quantile(0.75) - absences_data.absences.quantile(0.25)
absences_data = absences_data[absences_data.absences.between(perc25 - 1.5*IQR,
perc75 + 1.5*IQR)]

absences_data
```

```python
absences_data['bin'] = absences_data['absences'].apply(lambda x: 4*np.floor(x/4))
absences_data

absences_grouped = pd.DataFrame(absences_data.groupby(['bin']).mean())
absences_grouped
```

```python
absences_grouped['score'].plot()
# при достаточно больших выборках заметно сильный разброс результатов, нужна проверка на корреляцию
```
